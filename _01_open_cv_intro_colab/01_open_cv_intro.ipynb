```
import cv2
from google.colab.patches import cv2_imshow
```

### Read and Show

```
img_path = "pooya.jpg"
img = cv2.imread(img_path)

cv2_imshow(img)
```

### BGR2RGB

```
import matplotlib.pyplot as plt
```

```
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```
plt.imshow(img)
```

### Rectangle

```
cv2.rectangle(img, (200, 200), (500, 600), (255, 0 ,0), 2 )
cv2_imshow(img)
```

### Put Text

```
cv2.putText(img, 'pooya mohammadi', (200, 200), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0,0,255))
cv2_imshow(img)
```

### write

```
cv2.imwrite('pooya_mohammadi.png', img)
```
